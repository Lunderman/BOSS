# Experiment: 2 term truth with 1 moment
### Can be fit by a 1 term model

```python
N_Models = [1,2]

All_Samples = []
All_IACT = []
All_AIC = []
All_BIC = []

for nModel in N_Models:
    nTruth = 2
    nMoment = 1
    param_Truth = [ 2.0,    1.5,   0.9,  0.1]
    
    print('\n%s Term Model Fitting a %s Term Truth'%(nModel,nTruth))
    Res = Variable_Models.param_estimation(nTruth,nModel,nMoment,param_Truth, Plot_figs=True)
    All_Samples.append(Res[0])
    All_IACT.append(Res[1])
    All_AIC.append(Res[2])    
    All_BIC.append(Res[3])
    print('AIC = ', All_AIC[-1])
    print('BIC = ', All_BIC[-1])
    print('--------------------------------------------------------')
    print('--------------------------------------------------------')
    
```
