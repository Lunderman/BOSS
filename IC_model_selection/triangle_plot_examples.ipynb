```python
import numpy as np
import matplotlib.pyplot as plt
from triangle_plot import triplt
import matplotlib.cm as cm

plt.rcParams['font.size'] = 25
```

```python
nDim = 5
mu = [kk+1 for kk in range(nDim)]
C = np.random.randn(nDim,nDim)
C = C.T@C
X = np.random.multivariate_normal(mu,C,10000) 
```

```python
triplt(X,
       cmap_hist=cm.RdYlBu_r,
       linewidth= 5,
       fill_contours = True, 
       nLevels = 10, 
       bounds = None, 
       labels = ['Dim '+str(kk+1) for kk in range(nDim)],
       truth = mu, 
       truth_color='b',
       plot_mean = False,
       mean_color='r')
plt.show()
```

```python
triplt(X,
       cmap_hist=cm.RdYlBu_r,
       linewidth= 5,
       fill_contours = False, 
       nLevels = 10, 
       bounds = None, 
       labels = None,
       truth = None, 
       truth_color='b',
       plot_mean = True,
       mean_color='r')
plt.show()
```

```python
triplt(X,
       cmap_hist=cm.viridis,
       linewidth= 5,
       fill_contours = False, 
       nLevels = 50, 
       bounds = None, 
       labels = None,
       truth = None, 
       truth_color='b',
       plot_mean = True,
       mean_color='r')
plt.show()
```

```python
# Perturb the truth so the truth and sample mean do not align 
# this is only to illustrate plotting both the truth and sample mean
mu = [kk+1+np.random.randn() for kk in range(nDim)] 

triplt(X,
       cmap_hist=cm.gray_r,
       linewidth= 5,
       fill_contours = True, 
       nLevels = 10, 
       bounds = None, 
       labels = ['Dim '+str(kk+1) for kk in range(nDim)],
       truth = mu, 
       truth_color='b',
       plot_mean = True,
       mean_color='r')
plt.show()
```
